```python
import seaborn as sns
import matplotlib.pyplot as plt 
import numpy as np

def f(x):
    d=np.array([0.5,9/2])[np.newaxis][np.newaxis].T * np.ones_like(x)
    print(d.shape)
    return np.tensordot(np.array([0.5,9/2]), x**2, axes=([0],[0]))

from mpl_toolkits.mplot3d import Axes3D

# Create data for the plot
x1 = np.linspace(-1, 1, 100)
x2 = np.linspace(-1, 1, 100)
X1, X2 = np.meshgrid(x1, x2)
X = np.array([X1, X2])
Z = f(X)
# Create a 3D plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X1, X2, Z, cmap='viridis')

# Add level curves
ax.contour(X1, X2, Z, levels=10, offset=ax.get_zlim()[0], cmap='coolwarm')

# Set labels and title
ax.set_xlabel('x1')
ax.set_ylabel('x2')
ax.set_zlabel('f(x)')
ax.set_title('Function f(x) = x^2 @ [0.5, 9/2]')

# Show the plot
plt.show()
```
